```python
import pandas as pd
```

```python
#Upload data from 'https://www.zillow.com/research/data/'
janDF = pd.read_csv('/content/drive/MyDrive/NoteBookFIles/CitiesJanuaryData.csv')

#Check data to verify proper upload
print(janDF.columns)
```

```python
#Drop Irrelevant Columns
janDF.drop(columns = ['RegionType', 'StateName', 'RegionID'])
```

```python
#Filter Data by State and Metro

#Select State
ColoradoJan = janDF[janDF["State"].str.contains("CO")]

#Select Metro
DenAurLakeMetroSingleFamilyJan = ColoradoJan[ColoradoJan["Metro"].str.contains("Denver-Aurora-Lakewood, CO", na=False)]
```

```python
ColoradoJan
DenAurLakeMetroSingleFamilyJan
```

```python
#Calculate Percentage Changes
DenAurLakeMetroSingleFamilyJan['five_year_percentage_change'] = (DenAurLakeMetroSingleFamilyJan['2022-12-31'] - DenAurLakeMetroSingleFamilyJan['2017-12-31']) / DenAurLakeMetroSingleFamilyJan['2017-12-31'] * 100
DenAurLakeMetroSingleFamilyJan['one_year_percentage_change'] = (DenAurLakeMetroSingleFamilyJan['2022-12-31'] - DenAurLakeMetroSingleFamilyJan['2021-12-31']) / DenAurLakeMetroSingleFamilyJan['2021-12-31'] * 100
DenAurLakeMetroSingleFamilyJan['ten_year_percentage_change'] = (DenAurLakeMetroSingleFamilyJan['2022-12-31'] - DenAurLakeMetroSingleFamilyJan['2012-12-31']) / DenAurLakeMetroSingleFamilyJan['2012-12-31'] * 100
```
